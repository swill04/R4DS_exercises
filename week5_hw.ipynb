## 9.2.1 Exercises

1. Create a scatterplot of hwy vs. displ where the points are pink filled in triangles.

````r
ggplot(mpg, aes(x = hwy, y = displ)) +
    geom_point(shape = 17, color = 'pink')
````

2. Why did the following code not result in a plot with blue points?
```r
ggplot(mpg) + 
  geom_point(aes(x = displ, y = hwy, color = "blue"))
```

Answer: Because there isn't any separation from the aesthetics so color is being treated as a category to sort on. By adding a parenthesis and then continuing with color this makes the points blue.

````r
ggplot(mpg) +
    geom_point(aes(x = displ, y =hwy) , color = 'blue')
````

3. What does the stroke aesthetic do? What shapes does it work with? (Hint: use ?geom_point)

Answer: It modifies the width of the border of points. It works with point plots (circular data points).

4. What happens if you map an aesthetic to something other than a variable name, like aes(color = displ < 5)? Note, you’ll also need to specify x and y.

Answer: It colors it based on the condition you enter. It colors based on True or False scale.

````r
ggplot(mpg, aes(x = displ, y = hwy)) +
    geom_point(aes(color = displ < 5))
````

## 9.3.1 Exercises

1. What geom would you use to draw a line chart? A boxplot? A histogram? An area chart?

Answer: For line chart, use geom_line(). Boxplot use geom_boxplot(). Histogram use geom_histogram(). And for area chart use geom_area().

2. Earlier in this chapter we used show.legend without explaining it:
```r
ggplot(mpg, aes(x = displ, y = hwy)) +
  geom_smooth(aes(color = drv), show.legend = FALSE)
```
What does show.legend = FALSE do here? What happens if you remove it? Why do you think we used it earlier?

Answer: It removes the right hand legend that tells what the lines are colored by. If you remove show.legend = FALSE, it will show the legend as default. They did this to compare to the group function not including a legend so that it matched the other graphs.

3. What does the se argument to geom_smooth() do?

Answer: It displays the confidence interval around the smoothing. It's true by default.

4. Recreate the R code necessary to generate the following graphs. Note that wherever a categorical variable is used in the plot, it’s drv.

````r
ggplot(mpg, aes(x = displ, y = hwy)) +
    geom_point() +
    geom_smooth(se = FALSE)
````

````r
ggplot(mpg, aes(x = displ, y = hwy)) +
    geom_point() +
    geom_smooth(aes(group = drv), se = FALSE)
````

````r
ggplot(mpg, aes(x = displ, y = hwy, color = drv)) +
    geom_point() +
    geom_smooth(se = FALSE)
````

````r
ggplot(mpg, aes(x = displ, y = hwy)) +
    geom_point(aes(color = drv)) +
    geom_smooth(se = FALSE)
````

````r
ggplot(mpg, aes(x = displ, y = hwy)) +
    geom_point(aes(color = drv)) +
    geom_smooth(aes(linetype = drv), se = FALSE)
````

````r
ggplot(mpg, aes(x = displ, y = hwy)) +
    geom_point(aes(stroke = 2), color = 'white') +
    geom_point(aes(color = drv))
````
