```r
library(ggplot2)
```

```r
library(tidyverse)
```

```r
user2025 <- readr::read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/main/data/2025/2025-04-29/user2025.csv')
```

```r
user2025
```

```r
session_plot <- ggplot(user2025, aes(room, date)) +
    geom_point(aes(color = session, shape = video_recording)) +
    labs(
        title = 'Plot of presentations with the location and date',
        subtitle = 'Colored based on session type and shaped based on if it will be recorded or not'
    )
options(repr.plot.width = 15, repr.plot.height = 9)
session_plot
```
