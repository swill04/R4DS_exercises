10. Will these two graphs look different? Why/why not?
```r
ggplot(
  data = penguins,
  mapping = aes(x = flipper_length_mm, y = body_mass_g)
) +
  geom_point() +
  geom_smooth()

ggplot() +
  geom_point(
    data = penguins,
    mapping = aes(x = flipper_length_mm, y = body_mass_g)
  ) +
  geom_smooth(
    data = penguins,
    mapping = aes(x = flipper_length_mm, y = body_mass_g)
  )
```

Answer: These two graphs won't look different because the mapping aesthetics are the same even though the format of the plotting lines are different. In the first it combines what the geom_point() and geom_smooth() will use for variables. In the second, it explicitly states what each uses for it's input.

````r
ggplot(
  data = penguins,
  mapping = aes(x = flipper_length_mm, y = body_mass_g)
) +
  geom_point() +
  geom_smooth()

ggplot() +
  geom_point(
    data = penguins,
    mapping = aes(x = flipper_length_mm, y = body_mass_g)
  ) +
  geom_smooth(
    data = penguins,
    mapping = aes(x = flipper_length_mm, y = body_mass_g)
  )
````

2. How are the following two plots different? Which aesthetic, color or fill, is more useful for changing the color of bars?
```r
ggplot(penguins, aes(x = species)) +
  geom_bar(color = "red")

ggplot(penguins, aes(x = species)) +
  geom_bar(fill = "red")
```

Answer: The first only colors the outline of the bars whereas the second changes the outline as well as the entire area within the bar. Fill is a better aesthetic for changing the color of the bars because it's completely filled rather than being outlined.

````r
ggplot(penguins, aes(x = species)) +
  geom_bar(color = "red")

ggplot(penguins, aes(x = species)) +
  geom_bar(fill = "red")
````
